## **convert categorical columns**

```python
cat_cols = ['type','director','cast','rating','duration','listed_in','country']
df[cat_cols] = df[cat_cols].astype('category')
print(df[cat_cols])
```

## **cleaning Duration column**

```python
df[['duration_num','duration_unit']] = df['duration'].str.split(' ', expand=True)
df['duration_num'] = pd.to_numeric(df['duration_num'], errors='coerce')
print(df.head())
```

## **univariate analysis**

```python
plt.figure()
sns.displot(df['release_year'],bins=30,kde=True)
plt.title("Distribution of restaurant release year")
plt.xlabel("release_year")
plt.ylabel("density")
plt.show()
```

**count plot**

```python
plt.figure()
sns.countplot(x=df['type'])
plt.title("countplot  type")
plt.show()
```

**histogram plot**

```python
plt.figure()
sns.histplot(df['rating'],bins=30,kde=True)
plt.title("Distribution of restaurant rating")
plt.xlabel("rating")
plt.ylabel("release year")
plt.show()
```

**box plot**

```python
plt.figure()
sns.boxplot(y=df['release_year'],x=df['type'])
plt.xlabel='content_type'
plt.ylabel='duration'
plt.title('boxplot on content_type by duration')
```

*heat map**

```python
df['duration'] = df['duration'].astype(str)

df['duration'] = df['duration'].str.extract('(\d+)').astype(float)

corr=df[['release_year','duration']].corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
```

**pair plot**

```python
sns.pairplot(df[['release_year', 'duration']])
```
